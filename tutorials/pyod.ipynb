```python
import pyod
import importlib 
# List of model names
all_models = [
    'ABOD',
    #'AutoEncoder',
    'CBLOF',
    #'FeatureBagging',
    'HBOS',
    'IForest',
    'KNN',
    'LOF',
    'MCD',
    'OCSVM',
    'PCA'
]


# Dictionary to store the initialized models
model_instances = {}

# Loop through the list of model names
for model_name in all_models:
    try:
        # Create the module name by converting to lowercase (e.g., 'abod')
        module_name = model_name.lower()
        
        # Import the module dynamically from pyod.models
        module = importlib.import_module(f'pyod.models.{module_name}')
        
        # Get the class from the module (e.g., 'ABOD')
        ModelClass = getattr(module, model_name)
        
        # Initialize the model and store it in the dictionary
        model_instances[model_name] = ModelClass()
    except (ImportError, AttributeError) as e:
        print(f"Failed to import or initialize model {model_name}: {e}")
print(*model_instances.keys(), sep=', ')
```
